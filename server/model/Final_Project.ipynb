```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("Bengaluru_House_Data.csv")
df1.head()
```

```python
df2 = df1.drop(['area_type','availability','society','balcony'],axis='columns')
df2.head()
```

```python
df2.isnull().sum()
```

```python
df3 = df2.dropna()
df3.isnull().sum()
```

```python
df3.head()
```

```python
df3['bhk'] = df3['size'].apply(lambda x : int(x.split(' ')[0]))
```

```python
df3.head()
```

```python
df4 = df3.copy()
```

```python
def convert_to_float(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1]) ) / 2
    try:
        return float(x)
    except:
        return None
    
```

```python
df4['total_sqft'] = df3.total_sqft.apply(convert_to_float)
df4.head()
```

```python
df5 = df4.copy()
```

```python
df5.shape
```

```python
df5['price_per_sqft'] = df5['price'] * 100000 / df5['total_sqft']
```

```python
df5.head()
```

```python
len(df5.location.unique())
```

```python
df5.location = df5.location.apply(lambda x : x.strip())
```

```python
location_stats = df5.groupby('location')['location'].agg('count').sort_values(ascending = False)
location_stats
```

```python
location_less_than_10 = location_stats[location_stats <= 10]
location_less_than_10
```

```python
df5.location = df5.location.apply(lambda x : 'other' if x in location_less_than_10 else x)
```

```python
len(df5.location.unique())
```

```python
df5.shape
```

```python
df6 = df5[df5.total_sqft / df5.bhk >= 300]
df6.shape
```

```python
df6.head()
```

```python
def remove_pps_outlier(df):
    df_out = pd.DataFrame()
    for key , sub_df in df.groupby('location'):
        m = np.mean(sub_df.price_per_sqft)
        st = np.std(sub_df.price_per_sqft)
        reduced_df = sub_df[(sub_df.price_per_sqft > (m-st)) & (sub_df.price_per_sqft <= (m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
```

```python
df7 = remove_pps_outlier(df6)
```

```python
df7.shape
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])

    for location , location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk , bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean' : np.mean(bhk_df.price_per_sqft),
                'std' : np.std(bhk_df.price_per_sqft),
                'count' : bhk_df.shape[0]
            }
        for bhk , bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count'] > 5:
                exclude_indices = np.append(exclude_indices,bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
df8 = remove_bhk_outliers(df7)
df8.shape
```

```python
df9 = df8[df8.bath < df8.bhk+2]
df9.shape
```

```python
df9.head()
```

```python
df10 = df9.drop(['size','price_per_sqft'],axis='columns')
df10
```

```python
dummies = pd.get_dummies(df10.location)
dummies
```

```python
df10
```

```python
dummies = pd.get_dummies(df10.location)
dummies.head()
```

```python
dummies = dummies.replace([True,False],[1,0])
```

```python
dummies.head()
```

```python
df11 = pd.concat([df10,dummies],axis='columns')
```

```python
df11.head()
```

```python
df12 = df11.drop(['location','other'],axis='columns')
```

```python
df12.head()
```

```python
Y = df12.price
X = df12.drop(['price'],axis='columns')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
len(x_train)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(x_train,y_train)
lr_clf.score(x_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)
cross_val_score(LinearRegression(),X,Y,cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor


def find_best_model(X,Y):
    algos = {
        'linear_regression' : {
            'model' : LinearRegression(),
            'params' : {
            }
        },
        'lasso' : {
            'model' : Lasso(),
            'params' : {
                'alpha' : [1,2],
                'selection' : ['random','cyclic']
            }
        },
        'decision_tree' : {
            'model' : DecisionTreeRegressor(),
            'params' : {
                'criterion' : ['squared_error','friedman_mse'],
                'splitter' : ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)
    for algo_name , config in algos.items():
        gs = GridSearchCV(config['model'],config['params'],cv=cv,return_train_score=False)
        gs.fit(X,Y)
        scores.append({
            'model': algo_name,
            'best_score' : gs.best_score_,
            'best_params' : gs.best_params_
        })
    return pd.DataFrame(scores,columns=['model','best_score','best_params'])
```

```python
find_best_model(X,Y)
```

```python
def predict_price(location,sqft,bath,bhk):
    loc_index = np.where(X.columns == location)[0][0]

    x = np.zeros(len(X.columns))

    x[0] = sqft
    x[1] = bath
    x[2] = bhk

    if loc_index >= 0 :
        x[loc_index] = 1

    return lr_clf.predict([x])[0]
```

```python
predict_price('Indira Nagar',1500,2,4)
```
